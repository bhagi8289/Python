```
import pandas as pd
import numpy as np
```

```
max_temp_janjgir_pd = pd.read_csv('https://github.com/bhagi8289/datasets/raw/master/janjgir-max-data.csv',sep='\t') 
```

```
max_temp_janjgir_pd
```

```
# we have maximum temperature in degree Celsius for each month for 102 years
df = max_temp_janjgir_pd
df
```

```
# we see there is no missing values
df.isnull().count()
```

```
# we wish to calculate sample covariance matrix K for months temperatures
# https://en.wikipedia.org/wiki/Covariance
# here K = 12 i.e. 1 to 12 for each month
```

image.png

```
X = df[['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']].to_numpy()
```

```
X
```

```
X.shape
```

```
# the covariance matrix will be 12 x 12
# q11 q12 q13         ... q1,12
# q21 q22 q23         ... q2,12
# ...
# q12,1 q12,2 q12,3   ... q12,12
```

```
 K = np.zeros((12,12),dtype=np.float32)
 K
```

image.png

```
np.mean(X,axis=0)
```

```
for j in range(0,12):
  for k in range(0,12):
    for i in range(0,102):
      K[j,k] = K[j,k] + (X[i,j] - np.mean(X[:,j],axis=0))*(X[i,k] - np.mean(X[:,k],axis=0))
K = K /101
K 
```

```
print(np.var(X, axis=0, ddof=1))
```
